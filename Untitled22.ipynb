```python
posts = []

while True:
    print("\n1. Add Post")
    print("2. View Posts")
    print("3. Edit Post")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        title = input("Enter Blog Title: ")
        content = input("Enter Blog Content: ")

        posts.append({"title": title, "content": content})
        print("Post Published Successfully!")

    elif choice == "2":
        print("\n--- Published Posts ---")
        for i, post in enumerate(posts, 1):
            print(f"\nPost {i}")
            print("Title:", post["title"])
            print("Content:", post["content"])

    elif choice == "3":
        if len(posts) == 0:
            print("No posts available to edit!")
        else:
            print("\n--- Posts ---")
            for i, post in enumerate(posts, 1):
                print(f"{i}. {post['title']}")

            post_num = int(input("Enter post number to edit: "))

            if 1 <= post_num <= len(posts):
                new_title = input("Enter New Title: ")
                new_content = input("Enter New Content: ")

                posts[post_num - 1]["title"] = new_title
                posts[post_num - 1]["content"] = new_content

                print("Post Updated Successfully!")
            else:
                print("Invalid Post Number!")

    elif choice == "4":
        print("Thank you!")
        break

    else:
        print("Invalid Choice")
```
